```python
import csv
import os
from bs4 import BeautifulSoup

# Kullanıcıdan dosya adı veya klasör yolu almak
input_type = input("Dosya adı (tek dosya için) veya klasör yolu (birden fazla dosya için) girin: ").strip()

# Eğer kullanıcı bir klasör yolu girerse, klasördeki tüm HTML dosyalarını işle
if os.path.isdir(input_type):
    input_directory = input_type
    for filename in os.listdir(input_directory):
        if filename.endswith('.html'):
            input_file = os.path.join(input_directory, filename)
            
            # Dosya adını al, uzantı hariç
            output_file = os.path.splitext(filename)[0] + '.csv'
            output_file_path = os.path.join(input_directory, output_file)

            # HTML dosyasını oku
            with open(input_file, 'r', encoding='utf-8') as file:
                html_content = file.read()

            # HTML içeriğini BeautifulSoup ile parse et
            soup = BeautifulSoup(html_content, 'html.parser')

            # CSV dosyasını yazmak için gerekli olan bilgiler
            fields = ['Tür', 'Üniversite Adı']

            # CSV dosyasını açalım ve yazalım
            with open(output_file_path, mode='w', newline='', encoding='utf-8') as file:
                writer = csv.writer(file)
                writer.writerow(fields)  # Başlıkları yaz

                # HTML'deki tüm optgroup etiketlerini al
                for optgroup in soup.find_all('optgroup'):
                    university_type = optgroup['label']  # Optgroup etiketinin 'label' özelliği

                    # Her optgroup içindeki option etiketlerini al
                    for option in optgroup.find_all('option'):
                        university_name = option.text.strip()  # Üniversite adını al
                        writer.writerow([university_type, university_name])  # CSV'ye yaz

            print(f'{output_file} dosyası başarıyla oluşturuldu.')

# Eğer kullanıcı bir dosya adı girerse, sadece o dosyayı işle
elif os.path.isfile(input_type) and input_type.endswith('.html'):
    input_file = input_type
    
    # Dosya adını al, uzantı hariç
    output_file = os.path.splitext(os.path.basename(input_file))[0] + '.csv'
    output_file_path = os.path.join(os.path.dirname(input_file), output_file)

    # HTML dosyasını oku
    with open(input_file, 'r', encoding='utf-8') as file:
        html_content = file.read()

    # HTML içeriğini BeautifulSoup ile parse et
    soup = BeautifulSoup(html_content, 'html.parser')

    # CSV dosyasını yazmak için gerekli olan bilgiler
    fields = ['Tür', 'Üniversite Adı']

    # CSV dosyasını açalım ve yazalım
    with open(output_file_path, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(fields)  # Başlıkları yaz

        # HTML'deki tüm optgroup etiketlerini al
        for optgroup in soup.find_all('optgroup'):
            university_type = optgroup['label']  # Optgroup etiketinin 'label' özelliği

            # Her optgroup içindeki option etiketlerini al
            for option in optgroup.find_all('option'):
                university_name = option.text.strip()  # Üniversite adını al
                writer.writerow([university_type, university_name])  # CSV'ye yaz

    print(f'{output_file} dosyası başarıyla oluşturuldu.')

else:
    print("Geçersiz dosya veya klasör yolu girdiniz.")
```
